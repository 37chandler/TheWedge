### Task 2

* Files for this task: 

* Sample_of_Owners.ipynb
* owner_sample_results.csv

`Sample_of_Owners.ipynb`: 
This takes a random sample of card numbers from owners at the Wedge and then pulls all the records over time associated with their card number. This then writes the sample into a local file called `owner_sample_results.csv` which is around 250MB, a good sample size to test on. 
	

### Task 3

* Files for this task: 
* Sample_of_Owners.ipynb
* wedge_summary.db
* department_lookup.csv

`Building_Summary_Tables¶.ipynb`: 
This builds three different summary tables that describe sales by different dimensions. The tables are sales by date and hour, sales by year and month, and sales by product, month and year. To get the product description the file `department_lookup` was uploaded to Google Big Query to then join tables to find the summary. These summary tables are then stored in the `wedge_summary.db` that is a local database.

## Reflections
My experience of the Wedge project was difficult, time-consuming, and educational. I finished Task Two and Task Three, which gave me a little confidence to complete Task One. By no means, are Task Two and Three a walk in the park, but there’s some guidance from previous assignments. 
	<p>I realized that the main way to see if Task One worked, was by running the next two tasks and evaluating their output. Then, if it wasn’t close, I would go back and tweak Task One. However, each new upload of Task One took about 1 hour and 20 minutes. I would try to note what I had changed each time, because sometimes it would make it all worse. There were a few odd issues I tried my best to remedy, and I believe my final outputs are so close—besides the null value.</p>
	<p>Despite all the struggle, I feel like I have learned a lot about data engineering and processing data into a database. Which is huge. I am a big believer in learn by doing.  </p>
